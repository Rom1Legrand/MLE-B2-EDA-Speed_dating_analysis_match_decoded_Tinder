Premi√®re √©tape, √©tudier la distribution des matches selon les 6 principaux crit√®res : fun, ambition, intelligence, attractivit√©, sincerit√©, int√©r√™ts partag√©s

```python
# Taux de match global
match_rate = df['match'].mean()

# Distribution des matches par participant
matches_per_person = df.groupby('iid')['match'].mean()

# Corr√©lation entre les notes donn√©es et match
rating_cols = ['attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o']
correlations = df[rating_cols + ['match']].corr()['match'].sort_values(ascending=False)

print("Taux de match global:", match_rate)
print("\nCorr√©lations avec match:")
print(correlations)

# Visualisation des corr√©lations
plt.figure(figsize=(10,6))
sns.barplot(x=correlations.index, y=correlations.values)
plt.xticks(rotation=45)
plt.title('Corr√©lation entre notes donn√©es et matches')
```

```python
# Analyser les correlations de matches s√©par√©ment pour hommes et femmes
for gender in [0, 1]:
   gender_data = df[df['gender'] == gender]
   
   # Taux de match
   match_rate = gender_data['match'].mean()
   
   # Corr√©lations
   correlations = gender_data[rating_cols + ['match']].corr()['match'].sort_values(ascending=False)
   
   gender_label = "Femmes" if gender == 0 else "Hommes"
   print(f"\n{gender_label}:")
   print(f"Taux de match: {match_rate:.2%}")
   print("Corr√©lations avec match:")
   print(correlations)

# Visualisation comparative
plt.figure(figsize=(12,6))
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,6))

# Femmes
gender_data = df[df['gender'] == 0]
corr_f = gender_data[rating_cols + ['match']].corr()['match'].sort_values(ascending=False)
sns.barplot(x=corr_f.index, y=corr_f.values, ax=ax1)
ax1.set_title('Corr√©lations - Femmes')
ax1.tick_params(axis='x', rotation=45)

# Hommes
gender_data = df[df['gender'] == 1]
corr_h = gender_data[rating_cols + ['match']].corr()['match'].sort_values(ascending=False)
sns.barplot(x=corr_h.index, y=corr_h.values, ax=ax2)
ax2.set_title('Corr√©lations - Hommes')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
```

Mais il serait int√©ressant de regarder le taux de match (proportion) plut√¥t que le nombre absolu car m√™me si il y a plus de matches en nombre absolu √† 27 ans, cela pourrait simplement √™tre d√ª au fait qu'il y a plus de participants de cet √¢ge

Calculons le ratio matches/participants par √¢ge pour voir si certains √¢ges ont un "meilleur taux de r√©ussite"

```python
# Cr√©ation des tranches d'√¢ge
df['age_group'] = pd.cut(df['age'], 
                        bins=[18, 22, 25, 28, 31, 34, 50], 
                        labels=['18-22', '23-25', '26-28', '29-31', '32-34', '35+'])

# Calcul des statistiques par tranche d'√¢ge
stats_by_age = df.groupby('age_group').agg({
    'match': ['count', 'mean'],
    'age': 'count'
}).round(3)

stats_by_age.columns = ['Nombre de matches', 'Taux de match', 'Nombre total']
stats_by_age['Taux de match'] = (stats_by_age['Taux de match'] * 100).round(1)

# Affichage des r√©sultats
print("\nStatistiques par tranche d'√¢ge :")
print(stats_by_age)

# Cr√©ation d'un graphique des taux de match
plt.figure(figsize=(10, 6))
plt.bar(range(len(stats_by_age)), stats_by_age['Taux de match'], color='skyblue')
plt.title("Taux de match par tranche d'√¢ge")
plt.xlabel("Tranche d'√¢ge")
plt.ylabel("Taux de match (%)")
plt.xticks(range(len(stats_by_age)), stats_by_age.index, rotation=45)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

1. Les jeunes adultes performent mieux :
- Le taux de match le plus √©lev√© se trouve dans les tranches 18-22 ans et 23-25 ans (environ 17.5%)
- Il y a une l√©g√®re sup√©riorit√© pour la tranche 23-25 ans, qui repr√©sente le pic

2. D√©croissance progressive avec l'√¢ge :
- On observe une diminution r√©guli√®re du taux de match √† partir de 25 ans
- Entre 26-28 ans, le taux reste encore bon (environ 16%)
- La baisse s'accentue apr√®s 29 ans
- Le taux le plus bas est pour la tranche 32-34 ans (environ 11.5%)

3. L√©g√®re remont√©e apr√®s 35 ans :
- De fa√ßon surprenante, il y a une petite remont√©e pour les 35+ ans (environ 13%)
- Cependant, il faut noter que cette tranche d'√¢ge compte moins de participants, ce qui peut rendre la statistique moins fiable

4. Amplitude des variations :
- La diff√©rence entre le meilleur taux (23-25 ans √† ~17.5%) et le pire (32-34 ans √† ~11.5%) est d'environ 6 points
- Cette variation est significative mais pas extr√™me, sugg√©rant que l'√¢ge n'est pas le seul facteur d√©terminant dans le succ√®s des matches

Ces statistiques contredisent un peu la premi√®re analyse bas√©e sur les nombres absolus : m√™me si il y avait plus de matches en nombre absolu vers 25-27 ans, le taux de r√©ussite est en fait meilleur pour les plus jeunes utilisateurs.

```python
# Cr√©er la diff√©rence d'√¢ge
df['age_diff'] = df['age'] - df['age_o']

plt.rcParams.update({'font.size': 15})
fig, ax = plt.subplots(figsize=(12, 6))

# Distribution des matches selon la diff√©rence d'√¢ge - version c√¥te √† c√¥te
no_match = df[df['match'] == 0]['age_diff']
has_match = df[df['match'] == 1]['age_diff']

plt.hist([no_match, has_match], bins=20, range=(-15, 15), 
         label=['Pas de match', 'Match'], 
         color=['lightblue', 'orange'],
         alpha=0.7)

plt.title("Distribution des matches selon la diff√©rence d'√¢ge", pad=12, fontsize=14)
plt.xlabel("Diff√©rence d'√¢ge (√¢ge participant - √¢ge partenaire)")
plt.ylabel("Nombre de couples")
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()

# Statistiques sur les diff√©rences d'√¢ge dans les matches
match_data = df[df['match'] == 1]
print("\nDiff√©rence d'√¢ge moyenne dans les matches par genre:")
gender_means = match_data.groupby('gender')['age_diff'].mean()
print(gender_means.rename(index={0: 'Femmes', 1: 'Hommes'}).round(2))

# Analyse par cat√©gories
bins = [-float('inf'), -10, -5, -2, 0, 2, 5, 10, float('inf')]
labels = ['< -10 ans', '-10 √† -5 ans', '-5 √† -2 ans', '-2 √† 0 ans', 
          '0 √† 2 ans', '2 √† 5 ans', '5 √† 10 ans', '> 10 ans']
df['age_diff_cat'] = pd.cut(df['age_diff'], bins=bins, labels=labels)

# Taux de match par cat√©gorie et genre
match_rates = df.pivot_table(
    values='match',
    index='age_diff_cat',
    columns='gender',
    aggfunc='mean'
).round(3)

match_rates.columns = ['Femmes', 'Hommes']
print("\nTaux de match par cat√©gorie de diff√©rence d'√¢ge et par genre:")
print(match_rates * 100)  # Conversion en pourcentage
```

```python
# Cr√©er la figure et les axes
plt.rcParams.update({'font.size': 12})
fig, ax = plt.subplots(figsize=(15, 8))

# Cr√©er des donn√©es s√©par√©es pour match et non-match
no_match = df[df['match'] == 0]['age_diff']
has_match = df[df['match'] == 1]['age_diff']

# D√©finir les zones et leurs couleurs
zones = [(-15, -10), (-10, -5), (-5, -2), (-2, 0), (0, 2), (2, 5), (5, 10), (10, 15)]
zone_colors = ['#f0f0f0', '#e6e6e6', '#dcdcdc', '#d3d3d3', '#d3d3d3', '#dcdcdc', '#e6e6e6', '#f0f0f0']

# Ajouter les zones color√©es
for (start, end), color in zip(zones, zone_colors):
    ax.axvspan(start, end, color=color, alpha=0.3)

# Cr√©er l'histogramme
plt.hist([no_match, has_match], bins=30, range=(-15, 15), 
         label=['Pas de match', 'Match'], 
         color=['lightblue', 'orange'],
         alpha=0.7)

# Ajouter les annotations pour les zones
zone_labels = [
    'Tr√®s\ninf√©rieur\n(>10 ans)',
    'Nettement\ninf√©rieur\n(5-10 ans)',
    'L√©g√®rement\ninf√©rieur\n(2-5 ans)',
    'Proche\ninf√©rieur\n(0-2 ans)',
    'Proche\nsup√©rieur\n(0-2 ans)',
    'L√©g√®rement\nsup√©rieur\n(2-5 ans)',
    'Nettement\nsup√©rieur\n(5-10 ans)',
    'Tr√®s\nsup√©rieur\n(>10 ans)'
]

# Ajouter des lignes verticales pour les d√©limitations
for x in [-10, -5, -2, 0, 2, 5, 10]:
    plt.axvline(x=x, color='gray', linestyle='--', alpha=0.5)

# Ajouter les labels des zones
for i, ((start, end), label) in enumerate(zip(zones, zone_labels)):
    x_pos = (start + end) / 2
    y_pos = plt.ylim()[1] * 0.95  # Place le texte en haut
    plt.text(x_pos, y_pos, label, ha='center', va='top', fontsize=10)

plt.title("Distribution des matches selon la diff√©rence d'√¢ge", pad=40, fontsize=14)
plt.xlabel("Diff√©rence d'√¢ge (√¢ge participant - √¢ge partenaire)")
plt.ylabel("Nombre d'interactions")
plt.legend(loc='upper right')

# Ajuster les marges pour accommoder les labels
plt.tight_layout()

plt.show()

# Afficher les statistiques par zone
print("\nTaux de match par zone:")
print(match_rates * 100)  # Conversion en pourcentage
```

## LA RELIGION
---

```python
# S√©lectionner les matches
matched = df[df['match'] == 1]

# Grouper les niveaux d'importance
df['religion_level'] = pd.cut(df['imprelig'], 
                            bins=[0, 3, 7, 10], 
                            labels=['Faible', 'Moyenne', 'Forte'])

matches_by_importance = df[df['match'] == 1]['religion_level'].value_counts()
total_matches = len(matched)

print("Distribution des matches par importance religieuse (en %):")
print((matches_by_importance / total_matches * 100).round(2))

print("\nR√©sum√©:")
print(f"Faible: {(matches_by_importance['Faible']/total_matches*100):.1f}%")
print(f"Moyenne: {(matches_by_importance['Moyenne']/total_matches*100):.1f}%")
print(f"Forte: {(matches_by_importance['Forte']/total_matches*100):.1f}%")
```

Rappelons comment nous avons cat√©goris√© l'importance de la religion :
- Faible : notes de 0 √† 3
- Moyenne : notes de 4 √† 7
- Forte : notes de 8 √† 10

Les r√©sultats montrent que :
- 59.2% des matches sont entre personnes accordant une faible importance √† la religion (0-3)
- 27.6% des matches sont entre personnes accordant une importance moyenne (4-7)
- 13.2% des matches sont entre personnes accordant une forte importance (8-10)

Est ce que cela a une influence sur le match ?

```python
# 1. Importance d√©clar√©e de la religion (moyenne g√©n√©rale)
print("Importance moyenne d√©clar√©e de la religion:", df['imprelig'].mean())

# 2. Mais taux de match selon si m√™me importance religieuse ou non
df['same_religion_importance'] = (df['imprelig'] == df['imprelig']).astype(int)
print("\nTaux de match:")
print("M√™me importance religieuse:", df[df['same_religion_importance']==1]['match'].mean())
print("Importance religieuse diff√©rente:", df[df['same_religion_importance']==0]['match'].mean())
```

1. Les gens accordent une importance relativement faible √† la religion en moyenne

2. le taux de match est l√©g√®rement plus √©lev√© entre personnes ayant des niveaux d'importance religieuse diff√©rents qu'entre personnes ayant le m√™me niveau d'importance religieuse

- La religion est effectivement peu importante dans les d√©cisions de match
- Les diff√©rences d'importance religieuse ne semblent pas √™tre un frein au match
- Les gens semblent √™tre coh√©rents entre leur d√©claration (faible importance) et leur comportement


V√©rfions si cela est identique selon le sexe

```python
# Analyse par genre pour la religion
print("Importance moyenne d√©clar√©e de la religion:")
print(df.groupby('gender')['imprelig'].mean())

# Taux de match selon importance religieuse par genre
for gender in [0, 1]:  # 0=Femmes, 1=Hommes
    gender_df = df[df['gender'] == gender]
    print(f"\n{'Femmes' if gender == 0 else 'Hommes'}:")
    print("Importance religieuse d√©clar√©e:", gender_df['imprelig'].mean())
    
    # Taux de match si m√™me importance religieuse
    same_rel = gender_df[gender_df['same_religion_importance']==1]
    diff_rel = gender_df[gender_df['same_religion_importance']==0]
    print("Taux de match - m√™me importance religieuse:", same_rel['match'].mean())
    print("Taux de match - importance diff√©rente:", diff_rel['match'].mean())
```

Il y a des diff√©rences int√©ressantes pour la religion :

Les femmes accordent plus d'importance √† la religion que les hommes
mais paradoxalement :
pour les femmes : taux de match similaire que l'importance religieuse soit la m√™me ou diff√©rente
pour les hommes : matchent plus avec des personnes d'importance religieuse diff√©rente

Ces diff√©rences d'int√©r√™ts pourraient expliquer certains taux de match :

```python
# Pour chaque duo, nous devons comparer les scores d'activit√©s du participant avec ceux de son/sa partenaire
activities = ['sports', 'tvsports', 'exercise', 'dining', 'museums', 'art', 
              'hiking', 'gaming', 'clubbing', 'reading', 'tv', 'theater',
              'movies', 'concerts', 'music', 'shopping', 'yoga']

# Pour chaque activit√©, comparons les scores quand il y a match vs pas de match
for activity in activities:
    # Diff√©rence entre le score du participant et celui de son partenaire
    participant_score = df[activity]
    partner_score = df.merge(df[['iid', activity]], 
                           left_on='pid', 
                           right_on='iid', 
                           suffixes=('', '_partner'))[f'{activity}_partner']
    
    df[f'diff_{activity}'] = abs(participant_score - partner_score)

# Afficher les diff√©rences moyennes
print("Diff√©rences moyennes d'int√©r√™ts entre partenaires:")
for activity in activities:
    diff_col = f'diff_{activity}'
    print(f"\n{activity}:")
    print(f"Avec match: {df[df['match']==1][diff_col].mean():.2f}")
    print(f"Sans match: {df[df['match']==0][diff_col].mean():.2f}")
```
